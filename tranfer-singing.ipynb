```python
import os
from collections import defaultdict

# Đường dẫn input
base = "/kaggle/working/espnet/egs2/opencpop/svs2/data/stage1"
train_txt = "/kaggle/input/datasets/letunam/newdata/newdata/segments/train.txt"
test_txt = "/kaggle/input/datasets/letunam/newdata/newdata/segments/test.txt"

# Đường dẫn wavs
wavsdir = "/kaggle/input/datasets/letunam/newdata/newdata/wavs"

# Tạo thư mục output
for subset in ["train", "test"]:
    subset_dir = os.path.join(base, subset)
    os.makedirs(subset_dir, exist_ok=True)

def process_subset(txt_file, subset_dir):
    wavscp = os.path.join(subset_dir, "wav.scp")
    text = os.path.join(subset_dir, "text")
    utt2spk = os.path.join(subset_dir, "utt2spk")

    groups = defaultdict(list)

    with open(txt_file, "r", encoding="utf-8") as f_in:
        for line in f_in:
            parts = line.strip().split("|")
            if len(parts) < 2:
                continue
            uid = parts[0]
            lyric = parts[1]
            song_id = uid[:4]   # hoặc [:7] nếu muốn gom theo 7 ký tự đầu

            groups[song_id].append((uid, lyric))

    with open(wavscp, "w", encoding="utf-8") as f_wav, \
         open(text, "w", encoding="utf-8") as f_text, \
         open(utt2spk, "w", encoding="utf-8") as f_spk:

        for song_id, segs in sorted(groups.items()):
            # wav.scp: gom thành 1 entry cho cả bài
            # ở đây giả sử bạn có file wav gộp tên {song_id}.wav
            f_wav.write(f"{song_id} {wavsdir}/{song_id}.wav\n")

            # text: ghép lyric các đoạn
            full_lyric = " ".join([ly for _, ly in segs])
            f_text.write(f"{song_id} {full_lyric}\n")

            # utt2spk: tất cả gom về singer1
            f_spk.write(f"{song_id} singer1\n")

# Xử lý train và test
process_subset(train_txt, os.path.join(base, "train"))
process_subset(test_txt, os.path.join(base, "test"))

print("✅ Stage 1 đã tạo wav.scp, text, utt2spk cho train và test (gom theo bài)")
```

stage 2: chuẩn hóa các file từ stage 1 và tạo các file phụ

```python
import os
import shutil

base = "/kaggle/working/espnet/egs2/opencpop/svs2/data"
stage1 = os.path.join(base, "stage1")
stage2 = os.path.join(base, "stage2")
os.makedirs(stage2, exist_ok=True)

# Only train and test
subsets = ["train", "test"]

# Validate 'subsets' before use
if not isinstance(subsets, (list, tuple)):
    raise TypeError(f"'subsets' must be a list/tuple, got {type(subsets)}: {subsets}")

for dset in subsets:
    # Validate each item
    if not isinstance(dset, str):
        raise TypeError(f"Subset item must be str, got {type(dset)}: {dset}")
    if dset not in {"train", "test"}:
        raise ValueError(f"Unexpected subset value: {dset}")

    src_dir = os.path.join(stage1, dset)
    dst_dir = os.path.join(stage2, dset)
    print(f"→ Processing subset: {dset}")
    print(f"   src_dir={src_dir}")
    print(f"   dst_dir={dst_dir}")

    # Ensure stage1 subset exists (avoid creating junk in stage2 if source is wrong)
    if not os.path.isdir(src_dir):
        raise FileNotFoundError(f"Source subset directory not found: {src_dir}")

    os.makedirs(dst_dir, exist_ok=True)

    # Copy wav.scp, text, utt2spk from stage1 to stage2
    for fname in ["wav.scp", "text", "utt2spk"]:
        src = os.path.join(src_dir, fname)
        dst = os.path.join(dst_dir, fname)
        if os.path.exists(src):
            shutil.copy(src, dst)
            print(f"   ✅ Copied {fname}")
        else:
            print(f"   ⚠️ Missing {src}")

    # Write feats_type as a file, not a folder
    feats_type_path = os.path.join(dst_dir, "feats_type")
    with open(feats_type_path, "w") as f:
        f.write("raw\n")
    print(f"   ✅ Wrote feats_type at {feats_type_path}")

# 2) Create spk2sid and utt2sid
train_dir = os.path.join(stage2, "train")
utt2spk_file = os.path.join(train_dir, "utt2spk")
spk2sid_file = os.path.join(train_dir, "spk2sid")

if not os.path.exists(utt2spk_file):
    raise FileNotFoundError(f"utt2spk missing: {utt2spk_file}")

# Build spk2sid
speakers = set()
with open(utt2spk_file, "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) >= 2:
            speakers.add(parts[1])

with open(spk2sid_file, "w", encoding="utf-8") as f:
    f.write("<unk> 0\n")
    for i, spk in enumerate(sorted(speakers), start=1):
        f.write(f"{spk} {i}\n")
print(f"✅ Wrote spk2sid at {spk2sid_file}")

# Load spk->sid map
mapping = {}
with open(spk2sid_file, "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) == 2:
            mapping[parts[0]] = parts[1]

# Create utt2sid for train and test
for dset in ["train", "test"]:
    ddir = os.path.join(stage2, dset)
    utt2spk = os.path.join(ddir, "utt2spk")
    utt2sid = os.path.join(ddir, "utt2sid")
    if not os.path.exists(utt2spk):
        print(f"⚠️ Skip {dset}: missing {utt2spk}")
        continue

    with open(utt2spk, "r", encoding="utf-8") as f_in, open(utt2sid, "w", encoding="utf-8") as f_out:
        for line in f_in:
            parts = line.strip().split()
            if len(parts) >= 2:
                utt, spk = parts[0], parts[1]
                sid = mapping.get(spk, "0")
                f_out.write(f"{utt} {sid}\n")
    print(f"✅ Wrote utt2sid for {dset} at {utt2sid}")
# 3) Tạo spk2utt từ utt2spk
for dset in ["train", "test"]:
    ddir = os.path.join(stage2, dset)
    utt2spk = os.path.join(ddir, "utt2spk")
    spk2utt = os.path.join(ddir, "spk2utt")

    if os.path.exists(utt2spk):
        mapping = {}
        with open(utt2spk, "r", encoding="utf-8") as f:
            for line in f:
                utt, spk = line.strip().split()
                mapping.setdefault(spk, []).append(utt)

        with open(spk2utt, "w", encoding="utf-8") as f:
            for spk, utts in mapping.items():
                f.write(f"{spk} {' '.join(utts)}\n")

        print(f"✅ Tạo spk2utt cho {dset}: {len(mapping)} speakers")
    else:
        print(f"⚠️ Không tìm thấy {utt2spk} trong {ddir}")

print("🎉 Stage 2 completed: copied from stage1 and generated spk2sid/utt2sid")
```

```python
import os
import shutil
base = "/kaggle/working/espnet/egs2/opencpop/svs2/data"
stage2 = os.path.join(base, "stage2")
stage3 = os.path.join(base, "stage3")
os.makedirs(stage3, exist_ok=True)
# Thông số lọc
fs = 22050 # sampling rate Hz
min_wav_duration = 1.0 # giây
max_wav_duration = 500.0 # giây
min_length = int(min_wav_duration * fs)
max_length = int(max_wav_duration * fs)
def process_subset(dset):
    src_dir = os.path.join(stage2, dset)
    dst_dir = os.path.join(stage3, dset)
    os.makedirs(dst_dir, exist_ok=True)
    # Copy tất cả file metadata
    for fname in ["wav.scp", "text", "utt2spk", "spk2utt", "utt2sid", "feats_type", "utt2num_samples"]:
        src = os.path.join(src_dir, fname)
        if os.path.exists(src):
            shutil.copy(src, dst_dir)
    # Lọc text bỏ rỗng
    textfile = os.path.join(dst_dir, "text")
    if os.path.exists(textfile):
        with open(textfile) as f_in, open(textfile + ".tmp", "w") as f_out:
            for line in f_in:
                parts = line.strip().split(maxsplit=1)
                if len(parts) == 2 and parts[1].strip() != "":
                    f_out.write(line)
        os.replace(textfile + ".tmp", textfile)
    # Fix data dir: chỉ giữ utt xuất hiện trong tất cả file
    files = ["wav.scp", "text", "utt2spk"]
    utt_sets = []
    for fname in files:
        fpath = os.path.join(dst_dir, fname)
        if os.path.exists(fpath):
            utts = set(line.strip().split()[0] for line in open(fpath))
            utt_sets.append(utts)
    # Khởi tạo mặc định
    common_utts = set()
    if utt_sets:
        common_utts = set.intersection(*utt_sets)
        # Filter lại tất cả file
        for fname in files:
            fpath = os.path.join(dst_dir, fname)
            if os.path.exists(fpath):
                with open(fpath) as f_in, open(fpath + ".tmp", "w") as f_out:
                    for line in f_in:
                        utt = line.strip().split()[0]
                        if utt in common_utts:
                            f_out.write(line)
                os.replace(fpath + ".tmp", fpath)
    print(f"✅ Stage 3 đã xử lý {dset}: {len(common_utts)} utterances còn lại")
# Chỉ áp dụng cho train (và valid nếu có), không áp dụng cho test
for dset in ["train"]: # thêm "valid" nếu bạn có
    process_subset(dset)
```

dump/raw

```python
import os
import soundfile as sf
import numpy as np
import librosa

base = "/kaggle/working/espnet/egs2/opencpop/svs2"
stage3_train = os.path.join(base, "data/stage3/train")
stage3_test = os.path.join(base, "data/stage3_test/test")
dump_raw = os.path.join(base, "dump/raw")

# Tham số đặc trưng
fs = 22050
n_mels = 80
hop_length = 256

def dump_raw_features(src_dir, dst_dir):
    os.makedirs(os.path.join(dst_dir, "data"), exist_ok=True)

    wavscp = os.path.join(src_dir, "wav.scp")
    feats_scp = os.path.join(dst_dir, "feats.scp")
    utt2num_frames = os.path.join(dst_dir, "utt2num_frames")

    with open(wavscp) as f_in, \
         open(feats_scp, "w") as f_feats, \
         open(utt2num_frames, "w") as f_num:
        for line in f_in:
            utt, wav_path = line.strip().split()
            # Đọc wav
            wav, sr = sf.read(wav_path)
            if wav.ndim > 1:
                wav = wav.mean(axis=1)
            if sr != fs:
                wav = librosa.resample(wav, orig_sr=sr, target_sr=fs)
            # Tính Mel filterbank
            mel = librosa.feature.melspectrogram(
                y=wav,
                sr=fs,
                n_fft=1024,
                hop_length=hop_length,
                n_mels=n_mels,
                fmin=0,
                fmax=8000
            )
            logmel = np.log(np.maximum(mel, 1e-5))


            # Lưu đặc trưng
            feat_path = os.path.join(dst_dir, "data", f"{utt}.npy")
            logmel = logmel.T  # (T, 80)
            np.save(feat_path, logmel)
            # Ghi vào feats.scp và utt2num_frames
            f_num.write(f"{utt} {logmel.shape[0]}\n")
            f_feats.write(f"{utt} {feat_path}\n")

    print(f"✅ Dump raw features cho {src_dir} → {dst_dir}")

# Xử lý train và test
dump_raw_features(stage3_train, os.path.join(dump_raw, "train"))
dump_raw_features(stage3_test, os.path.join(dump_raw, "test"))
```

Stage 5: Chuẩn bị dữ liệu huấn luyện. Tạo token_list — danh sách các token (ký tự, âm vị, hoặc từ) được sử dụng trong huấn luyện mô hình text-based (ví dụ lyrics hoặc phoneme prediction).

```python
import os
from pathlib import Path

# === Cấu hình nguồn/đích ===
base = Path("/kaggle/working/espnet/egs2/opencpop/svs2")

# Stage3 inputs
train_dir = base / "data/stage3/train"
test_dir  = base / "data/stage3_test/test"

# Dump/raw outputs
dump_train = base / "dump/raw/train"
dump_test  = base / "dump/raw/test"

# === Hàm tạo srctexts từ utt2spk + text ===
def create_srctexts(src_dir: Path, out_dir: Path):
    utt2spk_path = src_dir / "utt2spk"
    text_path    = src_dir / "text"
    out_dir.mkdir(parents=True, exist_ok=True)
    srctexts_path = out_dir / "srctexts"

    # Kiểm tra đầu vào
    if not utt2spk_path.exists():
        raise FileNotFoundError(f"Không thấy: {utt2spk_path}")
    if not text_path.exists():
        raise FileNotFoundError(f"Không thấy: {text_path}")

    # Đọc danh sách utt từ utt2spk
    utts = []
    with open(utt2spk_path, "r", encoding="utf-8") as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) >= 1:
                utts.append(parts[0])

    # Đọc bản đồ utt → lyrics từ text
    utt2lyric = {}
    with open(text_path, "r", encoding="utf-8") as f:
        for line in f:
            parts = line.strip().split(maxsplit=1)
            if len(parts) == 2:
                utt, lyric = parts
                utt2lyric[utt] = lyric.strip()

    # Ghi srctexts: chỉ ghi các utt có lyrics
    written = 0
    missing = 0
    with open(srctexts_path, "w", encoding="utf-8") as out_f:
        for utt in utts:
            lyric = utt2lyric.get(utt, "")
            if lyric:
                out_f.write(f"{utt} {lyric}\n")
                written += 1
            else:
                missing += 1

    print(f"✅ Tạo xong: {srctexts_path}")
    print(f"   - Ghi {written} utt có lyrics")
    print(f"   - Bỏ qua {missing} utt thiếu lyrics trong file text")

# === Chạy cho train và test ===
create_srctexts(train_dir, dump_train)
create_srctexts(test_dir,  dump_test)
```

```python
import os
import sys
import subprocess
from pathlib import Path
from pypinyin import pinyin, Style

# ======================
# 0️⃣ Thêm ESPnet vào PYTHONPATH
# ======================
espnet_root = "/kaggle/working/espnet"
espnet2_path = os.path.join(espnet_root, "espnet2")

if not os.path.exists(espnet2_path):
    raise FileNotFoundError(f"❌ Không tìm thấy thư mục {espnet2_path}")

sys.path.append(espnet_root)
os.environ["PYTHONPATH"] = f"{espnet_root}:{os.environ.get('PYTHONPATH', '')}"
print(f"✅ Đã thêm {espnet_root} vào PYTHONPATH")

# ======================
# 1️⃣ Định nghĩa đường dẫn
# ======================
train_dir = Path("/kaggle/working/espnet/egs2/opencpop/svs2/data/stage3/train")
test_dir  = Path("/kaggle/working/espnet/egs2/opencpop/svs2/data/stage3_test/test")

dump_train = Path("/kaggle/working/espnet/egs2/opencpop/svs2/dump/raw/train")
dump_test  = Path("/kaggle/working/espnet/egs2/opencpop/svs2/dump/raw/test")

token_dir = Path("/kaggle/working/espnet/egs2/opencpop/svs2/dump/token_list/pinyin")
token_dir.mkdir(parents=True, exist_ok=True)
token_list = token_dir / "tokens.txt"

# ======================
# 2️⃣ Hàm chuyển text sang pinyin và tạo srctexts
# ======================
def create_srctexts(src_dir: Path, out_dir: Path):
    utt2spk_path = src_dir / "utt2spk"
    text_path    = src_dir / "text"
    out_dir.mkdir(parents=True, exist_ok=True)
    srctexts_path = out_dir / "srctexts"

    if not utt2spk_path.exists() or not text_path.exists():
        raise FileNotFoundError("❌ Thiếu utt2spk hoặc text trong " + str(src_dir))

    # Đọc danh sách utt
    utts = []
    with open(utt2spk_path, "r", encoding="utf-8") as f:
        for line in f:
            parts = line.strip().split()
            if parts:
                utts.append(parts[0])

    # Đọc lyrics
    utt2lyric = {}
    with open(text_path, "r", encoding="utf-8") as f:
        for line in f:
            parts = line.strip().split(maxsplit=1)
            if len(parts) == 2:
                utt, lyric = parts
                utt2lyric[utt] = lyric.strip()

    # Chuyển lyric sang pinyin
    written, missing = 0, 0
    with open(srctexts_path, "w", encoding="utf-8") as out_f:
        for utt in utts:
            lyric = utt2lyric.get(utt, "")
            if lyric:
                # chuyển sang pinyin (chuỗi phoneme cách nhau bằng khoảng trắng)
                pinyin_seq = " ".join([syll[0] for syll in pinyin(lyric, style=Style.TONE3)])
                out_f.write(f"{utt} {pinyin_seq}\n")
                written += 1
            else:
                missing += 1

    print(f"✅ Tạo srctexts tại {srctexts_path} (ghi {written}, bỏ {missing})")
    return srctexts_path

# Chạy cho train và test
srctexts_train = create_srctexts(train_dir, dump_train)
srctexts_test  = create_srctexts(test_dir, dump_test)

# ======================
# 3️⃣ Tạo tokens.txt từ srctexts (train)
# ======================
nlsyms_txt = token_dir / "nlsyms.txt"
if not nlsyms_txt.exists():
    open(nlsyms_txt, "w").close()

cmd = [
    "python3", "-m", "espnet2.bin.tokenize_text",
    "--token_type", "phn",   # vì đã chuyển sang phoneme (pinyin)
    "-f", "2-",
    "--input", str(srctexts_train),
    "--output", str(token_list),
    "--non_linguistic_symbols", str(nlsyms_txt),
    "--cleaner", "none",
    "--g2p", "none",         # không cần g2p nữa vì đã chuyển sang pinyin
    "--write_vocabulary", "true",
    "--add_symbol", "<blank>:0",
    "--add_symbol", "<unk>:1",
    "--add_symbol", "<sos/eos>:-1",
]

print("🚀 Đang sinh token_list từ:", srctexts_train)
result = subprocess.run(cmd, capture_output=True, text=True)
if result.returncode != 0:
    print("❌ Lỗi khi chạy tokenize_text:")
    print("stdout:\n", result.stdout)
    print("stderr:\n", result.stderr)
    raise subprocess.CalledProcessError(result.returncode, cmd)

print("✅ Đã tạo token_list tại:", token_list)

# ======================
# 4️⃣ Kiểm tra kết quả
# ======================
with open(token_list) as f:
    lines = f.read().strip().splitlines()
print(f"\n📜 Token list ({len(lines)} tokens):")
print("\n".join(lines[:30]), "\n...")
```

Stage 6 - F0 Ensemble + BigVGAN-Compatible Mel (FIXED)
========================================================
✓ 3 F0 estimators: librosa.pyin + pyworld.dio + pyworld.harvest
✓ Majority voting + Median ensemble
✓ Aggressive post-processing for octave errors
✓ **CRITICAL FIX**: Mel spectrogram matching BigVGAN exactly:
  - Power = 1.0 (magnitude spectrum, not power)
  - Mel norm = 'slaney' (matching BigVGAN)
  - Log scale = log10 (not natural log)
  - Audio normalization = 0.95 peak (matching BigVGAN training)

```python
import os
from pathlib import Path
import numpy as np
import soundfile as sf
import librosa
import pyworld as pw
from tqdm import tqdm
from scipy.ndimage import median_filter

# ===========================
# CONFIG
# ===========================
BASE = Path("/kaggle/working/espnet/egs2/opencpop/svs2")
train_wav_scp = BASE / "data/stage3/train/wav.scp"
valid_wav_scp = BASE / "data/stage3_test/test/wav.scp"

OUT_ROOT = Path("/kaggle/working/stage6")
OUT_ROOT.mkdir(parents=True, exist_ok=True)

# Audio params
TARGET_SR = 22050
N_FFT = 1024
HOP_LENGTH = 256
WIN_LENGTH = 1024
N_MELS = 80
FMIN, FMAX = 0, 8000

# F0 params
PYIN_FMIN = 80
PYIN_FMAX = 800
PYWORLD_F0_FLOOR = 80.0
PYWORLD_F0_CEIL = 800.0

# Valid F0 range
VALID_F0_MIN = 150
VALID_F0_MAX = 600

# Energy params
DB_FLOOR = -60

# BigVGAN audio normalization
AUDIO_NORM_SCALE = 0.95


# ===========================
# POST-PROCESSING
# ===========================
def post_process_f0(f0, voiced_flag):
    """
    Fix octave errors and outliers after ensemble
    """
    if voiced_flag.sum() < 10:
        return f0, voiced_flag
    
    voiced_f0 = f0[voiced_flag]
    median_f0 = np.median(voiced_f0)
    
    # 1. Detect octave errors
    if voiced_f0.max() > median_f0 * 2.0:
        print(f"    ⚠️  Detected octave errors (max={voiced_f0.max():.1f}Hz, median={median_f0:.1f}Hz)")
        high_mask = f0 > median_f0 * 1.8
        n_corrected = high_mask.sum()
        
        if n_corrected > 0:
            f0[high_mask] = f0[high_mask] / 2.0
            print(f"    ✓ Corrected {n_corrected} octave errors")
    
    # 2. Hard clip to valid range
    f0_before = f0.copy()
    f0 = np.clip(f0, 0, VALID_F0_MAX)
    
    n_clipped = ((f0_before > VALID_F0_MAX) & (f0_before > 0)).sum()
    if n_clipped > 0:
        print(f"    ✓ Clipped {n_clipped} values > {VALID_F0_MAX}Hz")
    
    # 3. Remove too-low values
    too_low = (f0 > 0) & (f0 < VALID_F0_MIN)
    n_removed = too_low.sum()
    
    if n_removed > 0:
        f0[too_low] = 0
        voiced_flag[too_low] = False
        print(f"    ✓ Removed {n_removed} values < {VALID_F0_MIN}Hz")
    
    # 4. Median filter
    if voiced_flag.sum() > 10:
        f0_smoothed = median_filter(f0, size=3, mode='nearest')#giam f0 smoothing
        f0 = np.where(voiced_flag, f0_smoothed, f0)
    
    return f0, voiced_flag


# ===========================
# F0 ENSEMBLE EXTRACTION
# ===========================
def extract_f0_ensemble(y, sr, utt_id):
    """Extract F0 using 3 estimators with ensemble"""
    
    # 1. PYIN
    try:
        f0_pyin, voiced_pyin, _ = librosa.pyin(
            y, fmin=PYIN_FMIN, fmax=PYIN_FMAX, sr=sr,
            frame_length=WIN_LENGTH, hop_length=HOP_LENGTH, center=False
        )
        if voiced_pyin is None:
            voiced_pyin = np.isfinite(f0_pyin) & (f0_pyin > 0)
        f0_pyin = np.nan_to_num(f0_pyin, nan=0.0)
        f0_pyin = np.where(voiced_pyin, f0_pyin, 0.0)
    except Exception as e:
        print(f"  ⚠️ PYIN failed for {utt_id}: {e}")
        f0_pyin = np.zeros(len(y) // HOP_LENGTH + 1)
        voiced_pyin = np.zeros(len(f0_pyin), dtype=bool)
    
    # 2. DIO
    try:
        _f0_dio, t_dio = pw.dio(
            y.astype(np.float64), sr,
            f0_floor=PYWORLD_F0_FLOOR, f0_ceil=PYWORLD_F0_CEIL,
            frame_period=HOP_LENGTH / sr * 1000
        )
        f0_dio = pw.stonemask(y.astype(np.float64), _f0_dio, t_dio, sr)
        voiced_dio = f0_dio > 0
    except Exception as e:
        print(f"  ⚠️ DIO failed for {utt_id}: {e}")
        f0_dio = np.zeros(len(y) // HOP_LENGTH + 1)
        voiced_dio = np.zeros(len(f0_dio), dtype=bool)
    
    # 3. HARVEST
    try:
        f0_harvest, t_harvest = pw.harvest(
            y.astype(np.float64), sr,
            f0_floor=PYWORLD_F0_FLOOR, f0_ceil=PYWORLD_F0_CEIL,
            frame_period=HOP_LENGTH / sr * 1000
        )
        voiced_harvest = f0_harvest > 0
    except Exception as e:
        print(f"  ⚠️ HARVEST failed for {utt_id}: {e}")
        f0_harvest = np.zeros(len(y) // HOP_LENGTH + 1)
        voiced_harvest = np.zeros(len(f0_harvest), dtype=bool)
    
    # 4. Align lengths
    target_len = len(y) // HOP_LENGTH + 1
    
    def align_length(arr, target):
        if len(arr) == target:
            return arr
        elif len(arr) > target:
            return arr[:target]
        else:
            return np.pad(arr, (0, target - len(arr)), constant_values=0)
    
    f0_pyin = align_length(f0_pyin, target_len)
    f0_dio = align_length(f0_dio, target_len)
    f0_harvest = align_length(f0_harvest, target_len)
    
    voiced_pyin = align_length(voiced_pyin, target_len)
    voiced_dio = align_length(voiced_dio, target_len)
    voiced_harvest = align_length(voiced_harvest, target_len)
    
    # 5. Majority voting
    vote_count = voiced_pyin.astype(int) + voiced_dio.astype(int) + voiced_harvest.astype(int)
    voiced_ensemble = vote_count >= 2
    
    # 6. Median F0
    f0_stack = np.stack([f0_pyin, f0_dio, f0_harvest], axis=0)
    f0_ensemble = np.zeros(target_len, dtype=np.float32)
    
    for i in range(target_len):
        if voiced_ensemble[i]:
            candidates = []
            if voiced_pyin[i] and f0_pyin[i] > 0:
                candidates.append(f0_pyin[i])
            if voiced_dio[i] and f0_dio[i] > 0:
                candidates.append(f0_dio[i])
            if voiced_harvest[i] and f0_harvest[i] > 0:
                candidates.append(f0_harvest[i])
            
            if len(candidates) > 0:
                f0_ensemble[i] = np.median(candidates)
            else:
                non_zero = f0_stack[:, i][f0_stack[:, i] > 0]
                if len(non_zero) > 0:
                    f0_ensemble[i] = np.median(non_zero)
                else:
                    f0_ensemble[i] = 0
                    voiced_ensemble[i] = False
    
    # 7. Post-processing (XÓA basic smoothing, chỉ gọi post_process_f0)
    f0_ensemble, voiced_ensemble = post_process_f0(f0_ensemble, voiced_ensemble)
    
    # 8. Logging
    n_voiced = voiced_ensemble.sum()
    if n_voiced > 0:
        voiced_f0 = f0_ensemble[f0_ensemble > 0]
        mean_f0 = voiced_f0.mean()
        median_f0 = np.median(voiced_f0)
        max_f0 = voiced_f0.max()
        min_f0 = voiced_f0.min()
        agreement = (vote_count == 3).sum() / len(vote_count) * 100
        
        print(f"  🎵 {utt_id}:")
        print(f"    Voiced: {n_voiced}/{target_len} ({n_voiced/target_len*100:.1f}%)")
        print(f"    F0: mean={mean_f0:.1f}Hz, median={median_f0:.1f}Hz")
        print(f"    F0 range: [{min_f0:.1f}, {max_f0:.1f}] Hz")
        print(f"    Estimator agreement: {agreement:.1f}%")
    
    return f0_ensemble, voiced_ensemble


# ===========================
# BIGVGAN-COMPATIBLE MEL EXTRACTION
# ===========================
def load_audio(wav_path):
    """Load audio and normalize like BigVGAN training"""
    y, sr = sf.read(str(wav_path), always_2d=False)
    if y.ndim == 2:
        y = y.mean(axis=1)
    y = y.astype(np.float32)
    
    if sr != TARGET_SR:
        y = librosa.resample(y, orig_sr=sr, target_sr=TARGET_SR)
    
    # **CRITICAL**: Normalize like BigVGAN training
    # BigVGAN uses: audio = librosa.util.normalize(audio) * 0.95
    y = librosa.util.normalize(y) * AUDIO_NORM_SCALE
    
    return y, TARGET_SR


def extract_log_mel_bigvgan(y, sr):
    """
    Extract mel spectrogram EXACTLY matching BigVGAN preprocessing
    
    BigVGAN uses:
    - power=1.0 (magnitude spectrum)
    - norm='slaney' for mel filterbank
    - log10 scale (not natural log)
    - center=False
    """
    
    # Compute mel spectrogram with Slaney norm
    S = librosa.feature.melspectrogram(
        y=y,
        sr=sr,
        n_fft=N_FFT,
        hop_length=HOP_LENGTH,
        win_length=WIN_LENGTH,
        center=False,  # BigVGAN uses center=False
        n_mels=N_MELS,
        fmin=FMIN,
        fmax=FMAX,
        power=1.0,      # ← CRITICAL: Magnitude spectrum (not power=2.0)
        norm='slaney',  # ← CRITICAL: Slaney normalization
        htk=False       # Use slaney mel scale
    )
    
    # AFTER (ĐÚNG):
    log_mel = np.log(np.maximum(S, 1e-5))
    
    return log_mel.T.astype(np.float32)


def extract_energy(y):
    """Extract RMS energy, normalized to [0,1]"""
    rms = librosa.feature.rms(
        y=y,
        frame_length=WIN_LENGTH,
        hop_length=HOP_LENGTH,
        center=False
    )[0]
    
    db = 20 * np.log10(np.maximum(rms, 1e-8))
    db = np.clip(db, DB_FLOOR, 0)
    energy = (db - DB_FLOOR) / (-DB_FLOOR)
    return energy.astype(np.float32)


def align_length(feat, target_len):
    """Align feature to target length"""
    cur_len = len(feat)
    if cur_len == target_len:
        return feat
    elif cur_len > target_len:
        return feat[:target_len]
    else:
        return np.pad(feat, (0, target_len - cur_len), constant_values=0)


# ===========================
# PROCESS SPLIT
# ===========================
def load_scp(scp_path):
    """Load SCP file"""
    lines = Path(scp_path).read_text().strip().splitlines()
    return [(l.split()[0], l.split()[1]) for l in lines if l.strip()]


def process_split(scp_path, split_name):
    """Process train/valid split"""
    pairs = load_scp(scp_path)
    print(f"\n{'='*70}")
    print(f"📌 {split_name}: {len(pairs)} files")
    print('='*70)
    
    features = []
    
    for utt_id, wav_path in tqdm(pairs, desc=split_name):
        try:
            # Load audio with BigVGAN normalization
            y, sr = load_audio(wav_path)
            
            # Extract features
            mel = extract_log_mel_bigvgan(y, sr)  # ← Using BigVGAN-compatible function
            f0, voiced = extract_f0_ensemble(y, sr, utt_id)
            energy = extract_energy(y)
            
            # Align lengths
            T = mel.shape[0]
            f0 = align_length(f0, T)
            energy = align_length(energy, T)
            
            # Save features
            feat_path = OUT_ROOT / f"{utt_id}_feats.npy"
            pitch_path = OUT_ROOT / f"{utt_id}_pitch.npy"
            energy_path = OUT_ROOT / f"{utt_id}_energy.npy"
            
            np.save(feat_path, mel)
            np.save(pitch_path, f0)
            np.save(energy_path, energy)
            
            features.append({
                'utt': utt_id,
                'mel': mel,
                'f0': f0,
                'energy': energy,
                'feat_path': feat_path,
                'pitch_path': pitch_path,
                'energy_path': energy_path
            })
            
        except Exception as e:
            print(f"❌ {utt_id}: {e}")
            continue
    
    print(f"✅ {split_name}: {len(features)} files processed")
    return features


# ===========================
# MAIN
# ===========================
if __name__ == "__main__":
    print("\n" + "="*70)
    print("🚀 STAGE 6 - F0 ENSEMBLE + BIGVGAN-COMPATIBLE MEL")
    print("="*70)
    print("✓ 3 Estimators: pyin + dio + harvest")
    print("✓ Majority voting + Median")
    print("✓ Aggressive post-processing")
    print(f"✓ Valid F0 range: {VALID_F0_MIN}-{VALID_F0_MAX} Hz")
    print("\n🔧 BigVGAN Compatibility:")
    print("✓ Audio normalization: librosa.normalize() * 0.95")
    print("✓ Mel power: 1.0 (magnitude spectrum)")
    print("✓ Mel norm: 'slaney'")
    print("✓ Log scale: log10")
    print("✓ Center: False")
    print("="*70)

    train_features = process_split(train_wav_scp, "TRAIN")
    valid_features = process_split(valid_wav_scp, "VALID")

    # Write SCP files
    train_feats_scp = OUT_ROOT / "train_feats.scp"
    train_pitch_scp = OUT_ROOT / "train_pitch.scp"
    train_energy_scp = OUT_ROOT / "train_energy.scp"
    valid_feats_scp = OUT_ROOT / "valid_feats.scp"
    valid_pitch_scp = OUT_ROOT / "valid_pitch.scp"
    valid_energy_scp = OUT_ROOT / "valid_energy.scp"

    with open(train_feats_scp, 'w') as ff, \
         open(train_pitch_scp, 'w') as fp, \
         open(train_energy_scp, 'w') as fe:
        for feat in train_features:
            ff.write(f"{feat['utt']} {feat['feat_path']}\n")
            fp.write(f"{feat['utt']} {feat['pitch_path']}\n")
            fe.write(f"{feat['utt']} {feat['energy_path']}\n")

    with open(valid_feats_scp, 'w') as ff, \
         open(valid_pitch_scp, 'w') as fp, \
         open(valid_energy_scp, 'w') as fe:
        for feat in valid_features:
            ff.write(f"{feat['utt']} {feat['feat_path']}\n")
            fp.write(f"{feat['utt']} {feat['pitch_path']}\n")
            fe.write(f"{feat['utt']} {feat['energy_path']}\n")

    # Compute stats
    print("\n" + "="*70)
    print("📊 Computing statistics...")
    print("="*70)

    all_mel = np.concatenate([f['mel'] for f in train_features + valid_features], axis=0)
    mel_mean = all_mel.mean(axis=0)
    mel_std = all_mel.std(axis=0)

    all_f0_voiced = []
    for f in train_features + valid_features:
        voiced = f['f0'][f['f0'] > 0]
        if len(voiced) > 0:
            all_f0_voiced.append(np.log(voiced + 1e-8))

    if all_f0_voiced:
        all_logf0 = np.concatenate(all_f0_voiced)
        f0_log_mean = all_logf0.mean()
        f0_log_std = all_logf0.std()
        
        all_f0_linear = np.concatenate([f['f0'][f['f0'] > 0] for f in train_features + valid_features])
        f0_linear_mean = all_f0_linear.mean()
        f0_linear_std = all_f0_linear.std()
        f0_linear_min = all_f0_linear.min()
        f0_linear_max = all_f0_linear.max()
        f0_linear_median = np.median(all_f0_linear)
    else:
        f0_log_mean = f0_log_std = 0.0
        f0_linear_mean = f0_linear_std = 0.0
        f0_linear_min = f0_linear_max = f0_linear_median = 0.0

    all_energy = np.concatenate([f['energy'] for f in train_features + valid_features], axis=0)
    energy_mean = all_energy.mean()
    energy_std = all_energy.std()

    stats_path = OUT_ROOT / "stats_full.npz"
    np.savez(
        stats_path,
        mel_mean=mel_mean.astype(np.float32),
        mel_std=mel_std.astype(np.float32),
        f0_log_mean=np.float32(f0_log_mean),
        f0_log_std=np.float32(f0_log_std),
        f0_linear_mean=np.float32(f0_linear_mean),
        f0_linear_std=np.float32(f0_linear_std),
        energy_mean=np.float32(energy_mean),
        energy_std=np.float32(energy_std)
    )

    print(f"\n✅ Stats saved: {stats_path}")
    print(f"   Mel: {mel_mean.mean():.4f} ± {mel_std.mean():.4f}")
    print(f"   F0 Log: {f0_log_mean:.4f} ± {f0_log_std:.4f}")
    print(f"   F0 Linear: {f0_linear_mean:.2f} Hz ± {f0_linear_std:.2f}")
    print(f"   F0 Range: [{f0_linear_min:.1f}, {f0_linear_max:.1f}] Hz")
    print(f"   F0 Median: {f0_linear_median:.1f} Hz")
    print(f"   Energy: {energy_mean:.4f} ± {energy_std:.4f}")

    print("\n" + "="*70)
    print("✅ STAGE 6 COMPLETE!")
    print("="*70)
    print("Expected improvements:")
    print("  - F0 max: 757Hz → 500-550Hz (octave error fixed)")
    print("  - GPE: 31.5% → 10-15% (-67%)")
    print("  - Voicing F1: 0.59 → 0.75-0.85")
    print("\n🎯 BigVGAN Compatibility:")
    print("  - Mel preprocessing now matches BigVGAN exactly")
    print("  - Should eliminate mel mismatch artifacts")
    print("  - Vocoder should decode cleanly now")
    print("="*70)
```

"""
Stage 7 - DiffSinger Training (ENHANCED VERSION)
=================================================
✓ Auxiliary decoder với L1 loss
✓ Shallow diffusion (configurable K)
✓ Multi-scale STFT loss
✓ More aggressive diffusion (SHALLOW_K=20)
✓ Expected: RMSE: 1.64 → 0.8-1.0 (-40-50%)
"""

```python
import os, gc, math, random
from pathlib import Path
import numpy as np
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

# ===========================
# CONFIG
# ===========================
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

# Audio params
SR = 22050
HOP_LENGTH = 256
N_MELS = 80
INPUT_DIM = N_MELS + 2

# Training params
BATCH_SIZE = 4
ACCUMULATION_STEPS = 6
NUM_WORKERS = 2

# Auxiliary decoder
AUX_EPOCHS = 200
AUX_LR = 1e-3

# Diffusion training
DIFF_EPOCHS = 250
DIFF_LR = 2e-4

# Diffusion params (MORE AGGRESSIVE)
DIFF_STEPS = 100
SHALLOW_K = 20  # Changed from 50 → More refinement steps
BETA_START = 0.0001
BETA_END = 0.02

CLIP_NORM = 1.0

# Crop params
CROP_SEC = 4.0
CROP_FRAMES = int(CROP_SEC * SR / HOP_LENGTH)

# Paths
FEAT_DIR = Path("/kaggle/working/stage6")
STATS_PATH = FEAT_DIR / "stats_full.npz"
OUTPUT_DIR = Path("/kaggle/working/stage7")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# ===========================
# LOAD STATS
# ===========================
print("\n📊 Loading normalization stats...")
stats = np.load(STATS_PATH)
MEL_MEAN = stats['mel_mean']
MEL_STD = stats['mel_std']
F0_LOG_MEAN = float(stats['f0_log_mean'])
F0_LOG_STD = float(stats['f0_log_std'])
EN_MEAN = float(stats['energy_mean'])
EN_STD = float(stats['energy_std'])
print(f"✅ Stats loaded")


# ===========================
# DATASET
# ===========================
def load_scp(path):
    d = {}
    with open(path, 'r') as f:
        for line in f:
            line = line.strip()
            if line:
                parts = line.split(maxsplit=1)
                if len(parts) == 2:
                    d[parts[0]] = parts[1]
    return d


class SVSDataset(Dataset):
    def __init__(self, feats_scp, pitch_scp, energy_scp, crop_frames=CROP_FRAMES):
        self.feats = load_scp(feats_scp)
        self.pitch = load_scp(pitch_scp)
        self.energy = load_scp(energy_scp)
        self.crop_frames = crop_frames
        
        self.utts = [u for u in self.feats if u in self.pitch and u in self.energy]
        print(f"Dataset: {len(self.utts)} utterances")
    
    def __len__(self):
        return len(self.utts)
    
    def _smart_crop(self, mel, f0, energy, crop_frames):
        T = mel.shape[0]
        
        if T <= crop_frames:
            pad = crop_frames - T
            mel = np.pad(mel, ((0, pad), (0, 0)), mode='constant')
            f0 = np.pad(f0, (0, pad), mode='constant')
            energy = np.pad(energy, (0, pad), mode='constant')
            return mel, f0, energy
        
        start = random.randint(0, T - crop_frames)
        end = start + crop_frames
        return mel[start:end], f0[start:end], energy[start:end]
    
    def __getitem__(self, idx):
        utt = self.utts[idx]
        
        mel = np.load(self.feats[utt]).astype(np.float32)
        f0 = np.load(self.pitch[utt]).astype(np.float32)
        energy = np.load(self.energy[utt]).astype(np.float32)
        
        T = mel.shape[0]
        
        if len(f0) != T:
            ratio = T / len(f0)
            f0 = np.repeat(f0, int(np.ceil(ratio)))[:T]
        
        if len(energy) != T:
            ratio = T / len(energy)
            energy = np.repeat(energy, int(np.ceil(ratio)))[:T]
        
        mel, f0, energy = self._smart_crop(mel, f0, energy, self.crop_frames)
        
        # Normalize
        mel_norm = (mel - MEL_MEAN) / (MEL_STD + 1e-8)
        mel_norm = np.clip(mel_norm, -5, 5)
        
        f0_log = np.where(f0 > 0, np.log(f0 + 1e-8), 0)
        f0_norm = np.where(f0 > 0, (f0_log - F0_LOG_MEAN) / (F0_LOG_STD + 1e-8), 0)
        f0_norm = np.clip(f0_norm, -5, 5)
        
        energy_norm = (energy - EN_MEAN) / (EN_STD + 1e-8)
        energy_norm = np.clip(energy_norm, -8, 8) 
        # Enhance contrast
        energy_norm = np.sign(energy_norm) * (np.abs(energy_norm) ** 1.2)
        energy_norm = np.clip(energy_norm, -5, 5)  # Clip lại sau khi enhance
        mel_t = torch.from_numpy(mel_norm).float()
        f0_t = torch.from_numpy(f0_norm).float().unsqueeze(-1)
        energy_t = torch.from_numpy(energy_norm).float().unsqueeze(-1)#khong norm ener
        
        x = torch.cat([mel_t, f0_t, energy_t], dim=-1)
        y = mel_t
        
        return x, y, self.crop_frames


def collate_fn(batch):
    xs, ys, lengths = zip(*batch)
    max_len = max(lengths)
    
    xs_pad, ys_pad = [], []
    for x, y in zip(xs, ys):
        T = x.shape[0]
        if T < max_len:
            x = F.pad(x, (0, 0, 0, max_len - T))
            y = F.pad(y, (0, 0, 0, max_len - T))
        xs_pad.append(x)
        ys_pad.append(y)
    
    xs_pad = torch.stack(xs_pad)
    ys_pad = torch.stack(ys_pad)
    lengths_t = torch.tensor(lengths, dtype=torch.long)
    
    return xs_pad, ys_pad, lengths_t


# ===========================
# MULTI-SCALE STFT LOSS (NEW)
# ===========================
class MultiScaleSTFTLoss(nn.Module):
    """
    Multi-scale STFT loss to capture high-frequency details
    
    Problem: MSE + MAE chỉ optimize pixel-wise difference
    → Không capture perceptual quality & high-freq details
    
    Solution: Compute loss at multiple time-frequency resolutions
    """
    def __init__(self):
        super().__init__()
        # Different resolutions
        self.fft_sizes = [512, 1024, 2048]
        self.hop_sizes = [128, 256, 512]
    
    def stft_magnitude(self, x, fft_size, hop_size):
        """
        Compute multi-scale representation
        x: (B, T, 80) mel
        """
        x = x.transpose(1, 2)  # (B, 80, T)
        
        # Different pooling for different scales
        if fft_size == 512:
            pool = nn.AvgPool1d(2, stride=2)
        elif fft_size == 1024:
            pool = nn.Identity()
        else:  # 2048
            pool = nn.Upsample(scale_factor=2, mode='linear', align_corners=False)
        
        x_pooled = pool(x)
        return x_pooled.transpose(1, 2)
    
    def forward(self, pred, target):
        """
        pred, target: (B, T, 80) mel spectrograms
        """
        loss = 0
        
        for fft, hop in zip(self.fft_sizes, self.hop_sizes):
            pred_stft = self.stft_magnitude(pred, fft, hop)
            target_stft = self.stft_magnitude(target, fft, hop)
            
            # Align lengths
            min_len = min(pred_stft.size(1), target_stft.size(1))
            pred_stft = pred_stft[:, :min_len]
            target_stft = target_stft[:, :min_len]
            
            # L1 loss at this scale
            loss += F.l1_loss(pred_stft, target_stft)
        
        return loss / len(self.fft_sizes)


# ===========================
# ENHANCED LOSS (NEW)
# ===========================
# ===========================
# ENHANCED LOSS - FIXED VERSION
# ===========================
class EnhancedLoss(nn.Module):
    def __init__(self, mel_mean, mel_std):
        super().__init__()
        self.register_buffer('mel_mean', torch.from_numpy(mel_mean).float())
        self.register_buffer('mel_std', torch.from_numpy(mel_std).float())
        
        self.mse = nn.MSELoss()
        self.l1 = nn.L1Loss()
        self.multi_scale = MultiScaleSTFTLoss()
    
    def high_freq_loss(self, pred, target):
        """Penalize error ở bins 50-80"""
        high_freq_pred = pred[:, :, 50:80]
        high_freq_target = target[:, :, 50:80]
        return F.l1_loss(high_freq_pred, high_freq_target)
    
    def mel_variation_loss(self, pred, target):
        """
        Penalize lack of temporal variation in mel
        Giữ vibrato gián tiếp qua mel variation
        """
        # Compute temporal gradient (difference between adjacent frames)
        pred_diff = torch.diff(pred, dim=1)      # (B, T-1, 80)
        target_diff = torch.diff(target, dim=1)  # (B, T-1, 80)
        
        # MSE on gradients
        return F.mse_loss(pred_diff, target_diff)
    def energy_envelope_loss(self, pred, target):
        """
        Penalize lack of attack/decay in energy envelope
        
        Attack: Tốc độ tăng năng lượng
        Decay: Tốc độ giảm năng lượng
        """
        # Compute energy per frame (RMS across mel bins)
        pred_energy = torch.sqrt(torch.mean(pred**2, dim=2))      # (B, T)
        target_energy = torch.sqrt(torch.mean(target**2, dim=2))  # (B, T)
        
        # Compute derivatives (attack/decay rate)
        pred_attack = torch.diff(pred_energy, dim=1)    # (B, T-1)
        target_attack = torch.diff(target_energy, dim=1)
        
        return F.mse_loss(pred_attack, target_attack)
        
    def forward(self, pred_norm, target_norm, f0_mask=None):
        """
        pred_norm, target_norm: (B, T, 80) normalized mel
        """
        # Denormalize for multi-scale loss
        pred = pred_norm * (self.mel_std + 1e-8) + self.mel_mean
        target = target_norm * (self.mel_std + 1e-8) + self.mel_mean
        
        # Component losses
        loss_mse = self.mse(pred, target)
        loss_mae = self.l1(pred, target)
        loss_ms = self.multi_scale(pred, target)
        loss_hf = self.high_freq_loss(pred, target)
        loss_var = self.mel_variation_loss(pred, target)  # ✅ SỬA: mel variation thay vì f0
        loss_env = self.energy_envelope_loss(pred, target)  
        
        # Combined with weights
        total = 0.5*loss_mse + 0.2*loss_mae + 0.8*loss_ms + 0.6*loss_hf + 0.3*loss_var+ 0.4*loss_env
        
        return total, {
            'mse': loss_mse.item(),
            'mae': loss_mae.item(),
            'ms': loss_ms.item(),
            'var': loss_var.item(), 
            'env': loss_env.item(), 
            'rmse': torch.sqrt(loss_mse).item()
        }

# ===========================
# AUXILIARY DECODER
# ===========================
class AuxiliaryDecoder(nn.Module):
    """FastSpeech2-like decoder"""
    def __init__(self, input_dim=82, hidden_dim=1024, num_layers=4):
        super().__init__()
        
        self.prenet = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Dropout(0.2)
        )
        
        self.rnn = nn.LSTM(
            256, hidden_dim, num_layers,
            batch_first=True, bidirectional=True,
            dropout=0.1 if num_layers > 1 else 0
        )
        
        self.postnet = nn.Sequential(
            nn.Linear(hidden_dim * 2, 512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, N_MELS)
        )
    
    def forward(self, x, lengths):
        x_pre = self.prenet(x)
        
        packed = nn.utils.rnn.pack_padded_sequence(
            x_pre, lengths.cpu(), batch_first=True, enforce_sorted=False
        )
        out, _ = self.rnn(packed)
        out, _ = nn.utils.rnn.pad_packed_sequence(out, batch_first=True)
        
        mel = self.postnet(out)
        return mel


# ===========================
# DIFFUSION MODEL
# ===========================
class GaussianDiffusion:
    """DDPM-based diffusion"""
    def __init__(self, beta_start=BETA_START, beta_end=BETA_END, timesteps=DIFF_STEPS, device=DEVICE):
        self.timesteps = timesteps
        self.device = device
        
        self.betas = torch.linspace(beta_start, beta_end, timesteps).to(device)
        self.alphas = 1.0 - self.betas
        self.alphas_cumprod = torch.cumprod(self.alphas, dim=0)
        self.alphas_cumprod_prev = F.pad(self.alphas_cumprod[:-1], (1, 0), value=1.0)
        
        self.sqrt_alphas_cumprod = torch.sqrt(self.alphas_cumprod)
        self.sqrt_one_minus_alphas_cumprod = torch.sqrt(1.0 - self.alphas_cumprod)
        
        self.posterior_variance = (
            self.betas * (1.0 - self.alphas_cumprod_prev) / (1.0 - self.alphas_cumprod)
        )
    
    def q_sample(self, x_start, t, noise=None):
        """Forward diffusion"""
        if noise is None:
            noise = torch.randn_like(x_start)
        
        sqrt_alphas_cumprod_t = self.sqrt_alphas_cumprod[t].view(-1, 1, 1)
        sqrt_one_minus_alphas_cumprod_t = self.sqrt_one_minus_alphas_cumprod[t].view(-1, 1, 1)
        
        return sqrt_alphas_cumprod_t * x_start + sqrt_one_minus_alphas_cumprod_t * noise


class DenoisingModel(nn.Module):
    """Denoising U-Net"""
    def __init__(self, mel_dim=80, cond_dim=82, hidden_dim=1024):
        super().__init__()
        
        self.time_embed = nn.Sequential(
            nn.Linear(128, hidden_dim),
            nn.SiLU(),
            nn.Linear(hidden_dim, hidden_dim)
        )
        
        self.cond_encoder = nn.Sequential(
            nn.Linear(cond_dim, 256),
            nn.ReLU(),
            nn.Linear(256, hidden_dim)
        )
        
        self.encoder = nn.ModuleList([
            nn.Sequential(
                nn.Conv1d(mel_dim, 256, 3, padding=1),
                nn.GroupNorm(8, 256),
                nn.SiLU()
            ),
            nn.Sequential(
                nn.Conv1d(256, 512, 3, padding=1),
                nn.GroupNorm(8, 512),
                nn.SiLU()
            )
        ])
        
        self.bottleneck = nn.Sequential(
            nn.Conv1d(512 + hidden_dim, 512, 3, padding=1),
            nn.GroupNorm(8, 512),
            nn.SiLU()
        )
        
        self.decoder = nn.ModuleList([
            nn.Sequential(
                nn.Conv1d(512 + 512, 256, 3, padding=1),
                nn.GroupNorm(8, 256),
                nn.SiLU()
            ),
            nn.Sequential(
                nn.Conv1d(256 + 256, mel_dim, 3, padding=1)
            )
        ])
    
    def time_embedding(self, t, dim=128):
        """Sinusoidal positional embedding"""
        half_dim = dim // 2
        emb = math.log(10000) / (half_dim - 1)
        emb = torch.exp(torch.arange(half_dim, device=t.device) * -emb)
        emb = t[:, None] * emb[None, :]
        emb = torch.cat([torch.sin(emb), torch.cos(emb)], dim=-1)
        return emb
    
    def forward(self, x, t, cond):
        """
        x: (B, T, 80) noisy mel
        t: (B,) timestep
        cond: (B, T, 82) condition
        """
        t_emb = self.time_embed(self.time_embedding(t))
        cond_enc = self.cond_encoder(cond)
        
        x = x.transpose(1, 2)
        
        skips = []
        for layer in self.encoder:
            x = layer(x)
            skips.append(x)
        
        t_emb_expanded = t_emb[:, :, None].expand(-1, -1, x.shape[2])
        cond_enc = cond_enc.transpose(1, 2)
        
        x = torch.cat([x, t_emb_expanded], dim=1)
        x = self.bottleneck(x)
        
        for i, layer in enumerate(self.decoder):
            if i < len(skips):
                x = torch.cat([x, skips[-(i+1)]], dim=1)
            x = layer(x)
        
        x = x.transpose(1, 2)
        return x


# ===========================
# TRAINING FUNCTIONS
# ===========================
def train_auxiliary(model, loader, optimizer, loss_fn, epoch, device):
    """Train auxiliary decoder with enhanced loss"""
    model.train()
    total_loss = 0
    total_mse = 0
    total_mae = 0
    total_ms = 0
    total_var = 0
    total_env = 0
    num_batches = 0
    
    pbar = tqdm(loader, desc=f"Aux Epoch {epoch}")
    for x, y, lengths in pbar:
        x = x.to(device)
        y = y.to(device)
        
        pred = model(x, lengths)
        pred = torch.clamp(pred, -5.0, 5.0)
        
        loss, metrics = loss_fn(pred, y)
        
        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), CLIP_NORM)
        optimizer.step()
        
        total_loss += loss.item()
        total_mse += metrics['mse']
        total_mae += metrics['mae']
        total_ms += metrics['ms']
        total_var += metrics['var'] 
        total_env += metrics['env']
        num_batches += 1
        
        pbar.set_postfix({
            'loss': f'{loss.item():.4f}',
            'rmse': f'{metrics["rmse"]:.4f}'
        })
    
    return {
        'loss': total_loss / num_batches,
        'mse': total_mse / num_batches,
        'mae': total_mae / num_batches,
        'ms': total_ms / num_batches,
        'var': total_var / num_batches, 
        'env': total_env / num_batches,  
        'rmse': np.sqrt(total_mse / num_batches)
    }

def train_diffusion(denoising_model, aux_model, diffusion, loader, optimizer, epoch, device):
    """Train diffusion model"""
    denoising_model.train()
    aux_model.eval()
    
    total_loss = 0
    num_batches = 0
    
    pbar = tqdm(loader, desc=f"Diff Epoch {epoch}")
    for x, y, lengths in pbar:
        x = x.to(device)
        y = y.to(device)
        
        with torch.no_grad():
            mel_coarse = aux_model(x, lengths)
            mel_coarse = torch.clamp(mel_coarse, -3.0, 3.0)
        
        # Sample from SHALLOW_K to DIFF_STEPS (more aggressive)
        t = torch.randint(SHALLOW_K, DIFF_STEPS, (x.shape[0],), device=device).long()
        
        noise = torch.randn_like(y)
        x_noisy = diffusion.q_sample(y, t, noise)
        noise_pred = denoising_model(x_noisy, t, x)
        
        loss = F.mse_loss(noise_pred, noise)
        
        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(denoising_model.parameters(), CLIP_NORM)
        optimizer.step()
        
        total_loss += loss.item()
        num_batches += 1
        
        pbar.set_postfix({'loss': f'{loss.item():.4f}'})
    
    return total_loss / num_batches


# ===========================
# MAIN
# ===========================
if __name__ == "__main__":
    print("\n" + "="*70)
    print("🚀 STAGE 7 - DIFFSINGER (ENHANCED)")
    print("="*70)
    print("✓ Multi-scale STFT loss")
    print(f"✓ Shallow diffusion: K={SHALLOW_K}/{DIFF_STEPS}")
    print("="*70)
    
    TRAIN_FEATS_SCP = FEAT_DIR / "train_feats.scp"
    TRAIN_PITCH_SCP = FEAT_DIR / "train_pitch.scp"
    TRAIN_ENERGY_SCP = FEAT_DIR / "train_energy.scp"
    
    print("\n📂 Loading datasets...")
    train_ds = SVSDataset(TRAIN_FEATS_SCP, TRAIN_PITCH_SCP, TRAIN_ENERGY_SCP)
    train_loader = DataLoader(
        train_ds, batch_size=BATCH_SIZE, shuffle=True,
        collate_fn=collate_fn, num_workers=NUM_WORKERS,
        pin_memory=True, drop_last=True
    )
    
    # ===== PHASE 1: AUXILIARY DECODER =====
    print("\n" + "="*70)
    print("PHASE 1: AUXILIARY DECODER")
    print("="*70)
    
    aux_model = AuxiliaryDecoder(input_dim=INPUT_DIM).to(DEVICE)
    aux_optimizer = torch.optim.AdamW(aux_model.parameters(), lr=AUX_LR)
    aux_loss_fn = EnhancedLoss(MEL_MEAN, MEL_STD).to(DEVICE)
    
    best_aux_rmse = float('inf')
    
    for epoch in range(1, AUX_EPOCHS + 1):
        metrics = train_auxiliary(aux_model, train_loader, aux_optimizer, aux_loss_fn, epoch, DEVICE)
        
        print(f"Aux Epoch {epoch}/{AUX_EPOCHS}:")
        print(f"  Loss={metrics['loss']:.4f}, RMSE={metrics['rmse']:.4f}")
        print(f"  MSE={metrics['mse']:.4f}, MAE={metrics['mae']:.4f}, MS={metrics['ms']:.4f}, Var={metrics['var']:.4f}, Env={metrics['env']:.4f}")
        
        if metrics['rmse'] < best_aux_rmse:
            best_aux_rmse = metrics['rmse']
            torch.save({
                'epoch': epoch,
                'model_state_dict': aux_model.state_dict(),
                'metrics': metrics
            }, OUTPUT_DIR / "aux_best.pt")
            print(f"  ✨ Best RMSE: {best_aux_rmse:.4f}")
        
        if epoch % 20 == 0:
            torch.save({
                'epoch': epoch,
                'model_state_dict': aux_model.state_dict()
            }, OUTPUT_DIR / f"aux_epoch_{epoch}.pt")
        
        gc.collect()
        torch.cuda.empty_cache()
    
    print(f"\n✅ Auxiliary training complete! Best RMSE: {best_aux_rmse:.4f}")
    
    # ===== PHASE 2: DIFFUSION =====
    print("\n" + "="*70)
    print("PHASE 2: DIFFUSION MODEL")
    print("="*70)
    
    ckpt = torch.load(OUTPUT_DIR / "aux_best.pt", weights_only=False)
    aux_model.load_state_dict(ckpt['model_state_dict'])
    aux_model.eval()
    
    diffusion = GaussianDiffusion(device=DEVICE)
    denoising_model = DenoisingModel().to(DEVICE)
    diff_optimizer = torch.optim.AdamW(denoising_model.parameters(), lr=DIFF_LR)
    
    best_diff_loss = float('inf')
    
    for epoch in range(1, DIFF_EPOCHS + 1):
        loss = train_diffusion(
            denoising_model, aux_model, diffusion,
            train_loader, diff_optimizer, epoch, DEVICE
        )
        
        print(f"Diff Epoch {epoch}/{DIFF_EPOCHS}: Loss={loss:.4f}")
        
        if loss < best_diff_loss:
            best_diff_loss = loss
            torch.save({
                'epoch': epoch,
                'denoising_model': denoising_model.state_dict(),
                'aux_model': aux_model.state_dict(),
                'loss': loss
            }, OUTPUT_DIR / "diffsinger_best.pt")
            print(f"  ✨ Best diff loss: {best_diff_loss:.4f}")
        
        if epoch % 30 == 0:
            torch.save({
                'epoch': epoch,
                'denoising_model': denoising_model.state_dict(),
                'aux_model': aux_model.state_dict()
            }, OUTPUT_DIR / f"diffsinger_epoch_{epoch}.pt")
        
        gc.collect()
        torch.cuda.empty_cache()
    
    print("\n" + "="*70)
    print("✅ DIFFSINGER COMPLETE!")
    print("="*70)
    print(f"Best aux RMSE: {best_aux_rmse:.4f}")
    print(f"Best diff loss: {best_diff_loss:.4f}")
    print("\nExpected improvements:")
    print("  - RMSE: 1.64 → 0.8-1.0 (-40-50%)")
    print("  - Multi-scale loss → Better high-freq")
    print(f"  - Shallow K={SHALLOW_K} → More refinement")
    print("="*70)
```

```python
#!/usr/bin/env python3
"""
Stage 7 - DiffSinger Inference
===============================
✓ Load trained auxiliary + denoising models
✓ Shallow diffusion sampling (from step K)
✓ Denormalize output for BigVGAN
"""

import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from pathlib import Path
from tqdm import tqdm
import math
from scipy.ndimage import gaussian_filter1d
# ===========================
# DEVICE & CONFIG
# ===========================
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"🔧 Using device: {DEVICE}")

SR = 22050
HOP_LENGTH = 256
N_MELS = 80
INPUT_DIM = N_MELS + 2

# Diffusion params (must match training)
DIFF_STEPS = 100
SHALLOW_K = 20
BETA_START = 0.0001
BETA_END = 0.02

# Paths
CHECKPOINT_PATH = Path("/kaggle/input/datasets/leenammta/checkpoint/diffsinger_best.pt")
FEAT_DIR = Path("/kaggle/input/datasets/leenammta/checkpoint/stage6")
STATS_PATH = FEAT_DIR / "stats_full.npz"
OUT_MEL_DIR = Path("/kaggle/working/stage7_pred")
OUT_MEL_DIR.mkdir(exist_ok=True, parents=True)
WAV_SCP = Path("/kaggle/working/espnet/egs2/opencpop/svs2/data/stage3_test/test/wav.scp")

# ===========================
# LOAD STATS
# ===========================
print("\n📊 Loading stats...")
stats = np.load(STATS_PATH)
MEL_MEAN = stats['mel_mean']
MEL_STD = stats['mel_std']
F0_LOG_MEAN = float(stats['f0_log_mean'])
F0_LOG_STD = float(stats['f0_log_std'])
EN_MEAN = float(stats['energy_mean'])
EN_STD = float(stats['energy_std'])

print("✅ Stats loaded")
print(f"📊 Normalization stats:")
print(f"   F0 log: mean={F0_LOG_MEAN:.4f}, std={F0_LOG_STD:.4f}")
print(f"   Energy: mean={EN_MEAN:.4f}, std={EN_STD:.4f}")
def add_f0_jitter(f0_norm, strength=0.02):
    """
    Thêm micro-variation vào F0 normalized
    
    Giọng người thật có:
    - Vibrato: ~5-7 Hz oscillation
    - Jitter: Random variation nhỏ
    """
    np.random.seed(None)  # Random mỗi lần
    
    # Chỉ jitter ở voiced frames
    voiced_mask = f0_norm != 0
    
    if voiced_mask.sum() < 10:
        return f0_norm
    
    # 1. Random jitter (±2%)
    jitter = np.random.normal(0, strength, len(f0_norm))
    jitter = gaussian_filter1d(jitter, sigma=2)  # Smooth
    
    # 2. Vibrato-like oscillation (optional, rất nhẹ)
    # t = np.arange(len(f0_norm))
    # vibrato = 0.01 * np.sin(2 * np.pi * 5.5 / HOP_LENGTH * t)
    # vibrato = gaussian_filter1d(vibrato, sigma=10)
    
    # Apply only to voiced
    f0_with_jitter = f0_norm.copy()
    f0_with_jitter[voiced_mask] += jitter[voiced_mask]
    # f0_with_jitter[voiced_mask] += vibrato[voiced_mask]  # Uncomment nếu muốn vibrato
    
    # Clip lại để không vượt quá range
    f0_with_jitter = np.clip(f0_with_jitter, -5, 5)
    
    return f0_with_jitter
    
# ===========================
# MODEL CLASSES (COPY EXACT FROM TRAINING)
# ===========================
class AuxiliaryDecoder(nn.Module):
    """FastSpeech2-like decoder"""
    def __init__(self, input_dim=82, hidden_dim=1024, num_layers=4):
        super().__init__()
        
        self.prenet = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Dropout(0.2)
        )
        
        self.rnn = nn.LSTM(
            256, hidden_dim, num_layers,
            batch_first=True, bidirectional=True,
            dropout=0.1 if num_layers > 1 else 0
        )
        
        self.postnet = nn.Sequential(
            nn.Linear(hidden_dim * 2, 512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, N_MELS)
        )
    
    def forward(self, x, lengths):
        x_pre = self.prenet(x)
        
        packed = nn.utils.rnn.pack_padded_sequence(
            x_pre, lengths.cpu(), batch_first=True, enforce_sorted=False
        )
        out, _ = self.rnn(packed)
        out, _ = nn.utils.rnn.pad_packed_sequence(out, batch_first=True)
        
        mel = self.postnet(out)
        return mel


class DenoisingModel(nn.Module):
    """Denoising U-Net"""
    def __init__(self, mel_dim=80, cond_dim=82, hidden_dim=1024):
        super().__init__()
        
        self.time_embed = nn.Sequential(
            nn.Linear(128, hidden_dim),
            nn.SiLU(),
            nn.Linear(hidden_dim, hidden_dim)
        )
        
        self.cond_encoder = nn.Sequential(
            nn.Linear(cond_dim, 256),
            nn.ReLU(),
            nn.Linear(256, hidden_dim)
        )
        
        self.encoder = nn.ModuleList([
            nn.Sequential(
                nn.Conv1d(mel_dim, 256, 3, padding=1),
                nn.GroupNorm(8, 256),
                nn.SiLU()
            ),
            nn.Sequential(
                nn.Conv1d(256, 512, 3, padding=1),
                nn.GroupNorm(8, 512),
                nn.SiLU()
            )
        ])
        
        self.bottleneck = nn.Sequential(
            nn.Conv1d(512 + hidden_dim, 512, 3, padding=1),
            nn.GroupNorm(8, 512),
            nn.SiLU()
        )
        
        self.decoder = nn.ModuleList([
            nn.Sequential(
                nn.Conv1d(512 + 512, 256, 3, padding=1),
                nn.GroupNorm(8, 256),
                nn.SiLU()
            ),
            nn.Sequential(
                nn.Conv1d(256 + 256, mel_dim, 3, padding=1)
            )
        ])
    
    def time_embedding(self, t, dim=128):
        """Sinusoidal positional embedding"""
        half_dim = dim // 2
        emb = math.log(10000) / (half_dim - 1)
        emb = torch.exp(torch.arange(half_dim, device=t.device) * -emb)
        emb = t[:, None] * emb[None, :]
        emb = torch.cat([torch.sin(emb), torch.cos(emb)], dim=-1)
        return emb
    
    def forward(self, x, t, cond):
        """
        x: (B, T, 80) noisy mel
        t: (B,) timestep
        cond: (B, T, 82) condition
        """
        t_emb = self.time_embed(self.time_embedding(t))
        cond_enc = self.cond_encoder(cond)
        
        x = x.transpose(1, 2)
        
        skips = []
        for layer in self.encoder:
            x = layer(x)
            skips.append(x)
        
        t_emb_expanded = t_emb[:, :, None].expand(-1, -1, x.shape[2])
        cond_enc = cond_enc.transpose(1, 2)
        
        x = torch.cat([x, t_emb_expanded], dim=1)
        x = self.bottleneck(x)
        
        for i, layer in enumerate(self.decoder):
            if i < len(skips):
                x = torch.cat([x, skips[-(i+1)]], dim=1)
            x = layer(x)
        
        x = x.transpose(1, 2)
        return x


class GaussianDiffusion:
    """DDPM-based diffusion"""
    def __init__(self, beta_start=BETA_START, beta_end=BETA_END, timesteps=DIFF_STEPS, device=DEVICE):
        self.timesteps = timesteps
        self.device = device
        
        self.betas = torch.linspace(beta_start, beta_end, timesteps).to(device)
        self.alphas = 1.0 - self.betas
        self.alphas_cumprod = torch.cumprod(self.alphas, dim=0)
        self.alphas_cumprod_prev = F.pad(self.alphas_cumprod[:-1], (1, 0), value=1.0)
        
        self.sqrt_alphas_cumprod = torch.sqrt(self.alphas_cumprod)
        self.sqrt_one_minus_alphas_cumprod = torch.sqrt(1.0 - self.alphas_cumprod)
        
        self.sqrt_recip_alphas = torch.sqrt(1.0 / self.alphas)
        self.sqrt_recipm1_alphas_cumprod = torch.sqrt(1.0 / self.alphas_cumprod - 1.0)
        
        self.posterior_variance = (
            self.betas * (1.0 - self.alphas_cumprod_prev) / (1.0 - self.alphas_cumprod)
        )
    
    @torch.no_grad()
    def p_sample(self, model, x, t, cond):
        """Single reverse diffusion step"""
        noise_pred = model(x, t, cond)
        
        betas_t = self.betas[t].view(-1, 1, 1)
        sqrt_one_minus_alphas_cumprod_t = self.sqrt_one_minus_alphas_cumprod[t].view(-1, 1, 1)
        sqrt_recip_alphas_t = self.sqrt_recip_alphas[t].view(-1, 1, 1)
        
        model_mean = sqrt_recip_alphas_t * (
            x - betas_t * noise_pred / sqrt_one_minus_alphas_cumprod_t
        )
        
        if t[0] > 0:
            noise = torch.randn_like(x)
            posterior_variance_t = self.posterior_variance[t].view(-1, 1, 1)
            return model_mean + torch.sqrt(posterior_variance_t) * noise
        else:
            return model_mean
    
    @torch.no_grad()
    def sample(self, model, mel_coarse, cond, start_step=SHALLOW_K):
        """Shallow diffusion sampling"""
        device = mel_coarse.device
        
        noise = torch.randn_like(mel_coarse)
        sqrt_alphas_cumprod_k = self.sqrt_alphas_cumprod[start_step]
        sqrt_one_minus_alphas_cumprod_k = self.sqrt_one_minus_alphas_cumprod[start_step]
        
        x = sqrt_alphas_cumprod_k * mel_coarse + sqrt_one_minus_alphas_cumprod_k * noise
        
        for t in reversed(range(start_step)):
            t_batch = torch.full((x.shape[0],), t, device=device, dtype=torch.long)
            x = self.p_sample(model, x, t_batch, cond)
        
        return x


# ===========================
# LOAD MODELS
# ===========================
print("\n🔨 Loading DiffSinger models...")

aux_model = AuxiliaryDecoder(input_dim=INPUT_DIM).to(DEVICE)
denoising_model = DenoisingModel().to(DEVICE)
diffusion = GaussianDiffusion(device=DEVICE)

if not CHECKPOINT_PATH.exists():
    raise FileNotFoundError(f"❌ Checkpoint not found: {CHECKPOINT_PATH}")

ckpt = torch.load(CHECKPOINT_PATH, map_location=DEVICE, weights_only=False)

aux_model.load_state_dict(ckpt['aux_model'])
denoising_model.load_state_dict(ckpt['denoising_model'])

aux_model.eval()
denoising_model.eval()

print(f"✅ Models loaded (epoch {ckpt.get('epoch', '?')})")
print(f"   Training loss: {ckpt.get('loss', 'N/A')}")


# ===========================
# LOAD VALID LIST
# ===========================
print("\n📂 Loading valid list...")
with open(WAV_SCP, "r") as f:
    utts = [l.split()[0] for l in f.readlines() if l.strip()]

print(f"   Total utterances: {len(utts)}")


# ===========================
# INFERENCE
# ===========================
print("\n" + "="*70)
print("🚀 STARTING DIFFSINGER INFERENCE")
print("="*70)

success_count = 0
fail_count = 0
all_f0_input = [] 
for utt in tqdm(utts, desc="Processing"):
    try:
        # Load features
        mel_path = FEAT_DIR / f"{utt}_feats.npy"
        f0_path = FEAT_DIR / f"{utt}_pitch.npy"
        en_path = FEAT_DIR / f"{utt}_energy.npy"
        
        if not (mel_path.exists() and f0_path.exists() and en_path.exists()):
            fail_count += 1
            continue
        
        mel = np.load(str(mel_path)).astype(np.float32)
        f0 = np.load(str(f0_path)).astype(np.float32)
        en = np.load(str(en_path)).astype(np.float32)
        
        # Fix shapes
        if mel.ndim == 2:
            if mel.shape[1] != N_MELS and mel.shape[0] == N_MELS:
                mel = mel.T
        
        if f0.ndim > 1:
            f0 = f0.flatten()
        if en.ndim > 1:
            en = en.flatten()
        
        T = mel.shape[0]
        
        if T < 10:
            fail_count += 1
            continue
        
        # Align lengths
        if len(f0) != T:
            ratio = T / len(f0)
            f0 = np.repeat(f0, int(np.ceil(ratio)))[:T]
        
        if len(en) != T:
            ratio = T / len(en)
            en = np.repeat(en, int(np.ceil(ratio)))[:T]
        
        # Normalize (SAME AS TRAINING)
        mel_norm = (mel - MEL_MEAN) / (MEL_STD + 1e-8)
        mel_norm = np.clip(mel_norm, -5, 5)
        
        f0_log = np.where(f0 > 0, np.log(f0 + 1e-8), 0)
        f0_norm = np.where(f0 > 0, (f0_log - F0_LOG_MEAN) / (F0_LOG_STD + 1e-8), 0)
        f0_norm = np.clip(f0_norm, -5, 5)
        f0_norm = add_f0_jitter(f0_norm, strength=0.015) 
        energy_norm = (en - EN_MEAN) / (EN_STD + 1e-8)
        energy_norm = np.clip(energy_norm, -8, 8)
        # Enhance contrast
        energy_norm = np.sign(energy_norm) * (np.abs(energy_norm) ** 1.2)
        energy_norm = np.clip(energy_norm, -5, 5)  # Clip lại sau khi enhance
        if success_count == 0:
            voiced_mask = f0 > 0
            if voiced_mask.sum() > 0:
                print(f"\n🔍 DEBUG - First file ({utt}):")
                print(f"   F0 input Hz: mean={f0[voiced_mask].mean():.1f}, "
                      f"min={f0[voiced_mask].min():.1f}, "
                      f"max={f0[voiced_mask].max():.1f}")
                print(f"   F0 normalized: mean={f0_norm[voiced_mask].mean():.3f}, "
                      f"min={f0_norm[voiced_mask].min():.3f}, "
                      f"max={f0_norm[voiced_mask].max():.3f}")
                print(f"   Voiced frames: {voiced_mask.sum()}/{len(f0)} "
                      f"({voiced_mask.sum()/len(f0)*100:.1f}%)")
        voiced_f0 = f0[f0 > 0]
        if len(voiced_f0) > 0:
            all_f0_input.extend(voiced_f0.tolist())
        # Concatenate input (SAME AS TRAINING)
        inp = np.concatenate([
            mel_norm,
            f0_norm.reshape(-1, 1),
            energy_norm.reshape(-1, 1)
        ], axis=1)
        
        # To tensor
        inp_t = torch.from_numpy(inp).unsqueeze(0).to(DEVICE)
        lengths = torch.tensor([T], dtype=torch.long)
        
        # ===== TWO-STAGE INFERENCE =====
        with torch.no_grad():
            # Stage 1: Auxiliary decoder
            mel_coarse = aux_model(inp_t, lengths)
            mel_coarse = torch.clamp(mel_coarse, -5.0, 5.0)
            
            # Stage 2: Diffusion refinement
            mel_refined = diffusion.sample(
                denoising_model,
                mel_coarse,
                inp_t,
                start_step=SHALLOW_K
            )
            mel_refined = torch.clamp(mel_refined, -5.0, 5.0)
        
        mel_pred_norm = mel_refined.squeeze(0).cpu().numpy()
        mel_pred_norm[:, 60:70] += 0.15 * mel_pred_norm[:, 60:70].std()
        mel_pred_norm[:, 70:80] += 0.25 * mel_pred_norm[:, 70:80].std()
        
        # Denormalize
        pred_mel = mel_pred_norm * (MEL_STD + 1e-8) + MEL_MEAN
        
        # Save
        out_path = OUT_MEL_DIR / f"{utt}.npy"
        np.save(out_path, pred_mel.astype(np.float32))
        
        success_count += 1
        
    except Exception as e:
        print(f"\n❌ Error {utt}: {e}")
        fail_count += 1


# ===========================
# SUMMARY
# ===========================
print("\n" + "="*70)
print("🎬 DIFFSINGER INFERENCE COMPLETED!")
print("="*70)
print(f"✅ Success: {success_count}/{len(utts)}")
print(f"❌ Failed: {fail_count}/{len(utts)}")

if len(all_f0_input) > 0:
    all_f0_input = np.array(all_f0_input)
    
    print(f"\n📊 F0 INPUT STATISTICS:")
    print(f"   Total voiced frames: {len(all_f0_input):,}")
    print(f"   F0 Hz: mean={all_f0_input.mean():.1f}, "
          f"median={np.median(all_f0_input):.1f}")
    print(f"   F0 Hz: min={all_f0_input.min():.1f}, "
          f"max={all_f0_input.max():.1f}, "
          f"std={all_f0_input.std():.1f}")
    
    # Compute log stats to compare with normalization
    f0_log_input = np.log(all_f0_input)
    print(f"\n📊 F0 LOG STATISTICS (should match training):")
    print(f"   Mean: {f0_log_input.mean():.4f} "
          f"(training: {F0_LOG_MEAN:.4f})")
    print(f"   Std:  {f0_log_input.std():.4f} "
          f"(training: {F0_LOG_STD:.4f})")
    
    # Check for bias
    diff_mean = abs(f0_log_input.mean() - F0_LOG_MEAN)
    diff_std = abs(f0_log_input.std() - F0_LOG_STD)
    
    if diff_mean > 0.1 or diff_std > 0.1:
        print(f"\n⚠️  WARNING: Stats mismatch detected!")
        print(f"   Mean diff: {diff_mean:.4f}")
        print(f"   Std diff:  {diff_std:.4f}")
        print(f"   This may cause F0 bias in output!")

print(f"\n📁 Output: {OUT_MEL_DIR}")
# Validation
if success_count > 0:
    sample_files = list(OUT_MEL_DIR.glob("*.npy"))
    sample = np.load(sample_files[0])
    
    print(f"\n📊 Sample output:")
    print(f"   Shape: {sample.shape}")
    print(f"   Range: [{sample.min():.2f}, {sample.max():.2f}]")
    print(f"   Mean: {sample.mean():.2f}")
    print(f"   Std: {sample.std():.2f}")
    
    if -15 < sample.mean() < 0 and 1 < sample.std() < 5:
        print(f"   ✅ Output looks valid!")

print("\n" + "="*70)
print("✅ READY FOR BIGVGAN VOCODER (Stage 8)")
print("="*70)
```

```python
#!/usr/bin/env python3
"""
Stage 8 - BigVGAN Vocoder Inference (CORRECTED)
================================================
✓ Load mel từ Stage 7 (ĐÃ denormalized)
✓ KHÔNG denormalize lại
✓ BigVGAN inference: mel → waveform
"""

import torch
import numpy as np
import soundfile as sf
from pathlib import Path
from tqdm import tqdm
from scipy.signal import butter, filtfilt
import torchaudio
from scipy.ndimage import gaussian_filter1d
# ===========================
# CONFIG
# ===========================
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"🔧 Using device: {DEVICE}")

SR = 22050
N_MELS = 80

# POST-FILTER CONFIG
ENABLE_POST_FILTER = True
CUTOFF_FREQ = 7500  # Hz - Gạt bỏ trên 7kHz
FILTER_ORDER = 3 

# ===========================
# POST-FILTER FUNCTION
# ===========================
def post_filter_audio(audio, sr=22050, cutoff=7000, order=4):
    """
    Low-pass filter để loại bỏ high-freq artifacts
    
    Args:
        audio: (samples,) numpy array
        sr: sample rate
        cutoff: cutoff frequency in Hz
        order: filter order (higher = steeper rolloff)
    
    Returns:
        filtered audio
    """
    # Normalize cutoff to Nyquist frequency
    nyquist = sr / 2.0
    normal_cutoff = cutoff / nyquist
    
    # Design Butterworth low-pass filter
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    
    # Apply zero-phase filter (forward + backward)
    filtered = filtfilt(b, a, audio)
    
    return filtered.astype(np.float32)
def add_natural_artifacts(audio, sr=22050):
    """
    Thêm các artifacts tự nhiên của giọng hát
    
    1. Breathiness: Pink noise nhẹ
    2. Micro-variations: Jitter/shimmer
    3. Transient enhancement: Làm nổi các âm đầu
    """
    audio = audio.copy()
    
    # 1. BREATHINESS - Thêm pink noise rất nhẹ
    # Giọng hát thật có breath noise từ luồng khí
    np.random.seed(42)  # Reproducible
    noise = np.random.randn(len(audio))
    
    # Pink noise filter (1/f spectrum)
    from scipy.signal import lfilter
    b = [0.049922035, -0.095993537, 0.050612699, -0.004408786]
    a = [1, -2.494956002, 2.017265875, -0.522189400]
    pink_noise = lfilter(b, a, noise)
    pink_noise = pink_noise / np.max(np.abs(pink_noise))
    
    # Chỉ thêm breath vào voiced segments (có năng lượng)
    rms = np.sqrt(np.convolve(audio**2, np.ones(512)/512, mode='same'))
    breath_mask = rms > 0.02  # Chỉ thêm khi có âm thanh
    
    breath_strength = 0.003  # RẤT NHẸ (0.3%)
    audio = audio + breath_mask * breath_strength * pink_noise
    
    
    # 2. MICRO-VARIATIONS - Jitter/shimmer rất nhẹ
    # Giọng người thật không bao giờ hoàn toàn stable
    # Tạo variation curve rất mịn
    variation_freq = 5  # Hz - tần số variation
    t = np.arange(len(audio)) / sr
    
    # Amplitude shimmer (±0.5%)
    shimmer = 1.0 + 0.005 * np.sin(2 * np.pi * variation_freq * t + np.random.rand())
    shimmer = gaussian_filter1d(shimmer, sigma=sr//10)  # Làm mượt
    
    audio = audio * shimmer
    
    
    # 3. TRANSIENT ENHANCEMENT - Làm nổi âm đầu
    # Tìm transients (đột biến năng lượng)
    audio_diff = np.diff(np.abs(audio), prepend=0)
    transient_mask = audio_diff > np.percentile(audio_diff, 99)
    
    # Boost nhẹ các transient
    boost = np.ones_like(audio)
    boost[transient_mask] = 1.08  # +8% tại transient
    boost = gaussian_filter1d(boost, sigma=5)  # Smooth transition
    
    audio = audio * boost
    
    return audio.astype(np.float32)  
# ===========================
# PATHS
# ===========================
INPUT_MEL_DIR = Path("/kaggle/working/stage7_pred")
if not INPUT_MEL_DIR.exists():
    raise FileNotFoundError(f"❌ Stage 7 output not found: {INPUT_MEL_DIR}")

OUTPUT_DIR = Path("/kaggle/working/stage8_audio")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# ===========================
# LOAD BIGVGAN
# ===========================
print("\n🚀 Loading BigVGAN vocoder...")

from bigvgan import BigVGAN

vocoder = BigVGAN.from_pretrained(
    "nvidia/bigvgan_22khz_80band",
    use_cuda_kernel=False
)
vocoder = vocoder.to(DEVICE)
vocoder.eval()

print(f"✅ BigVGAN loaded")
if ENABLE_POST_FILTER:
    print(f"✅ Post-filter enabled: cutoff={CUTOFF_FREQ}Hz, order={FILTER_ORDER}")



# ===========================
# PROCESS FILES
# ===========================
print("\n" + "="*70)
print("🎵 STARTING VOCODER + POST-FILTER")
print("="*70)

mel_files = sorted(INPUT_MEL_DIR.glob("*.npy"))
print(f"Found {len(mel_files)} mel files\n")

success_count = 0
fail_count = 0

for mel_path in tqdm(mel_files, desc="Generating audio"):
    utt_id = mel_path.stem
    output_path = OUTPUT_DIR / f"{utt_id}.wav"
    
    try:
        # Load mel
        mel = np.load(mel_path).astype(np.float32)  # (T, 80)
        
        # Validate & transpose
        if mel.ndim != 2:
            raise ValueError(f"Invalid mel shape: {mel.shape}")
        
        if mel.shape[1] != N_MELS:
            if mel.shape[0] == N_MELS:
                mel = mel.T
            else:
                raise ValueError(f"Wrong mel dimension: {mel.shape}")
        
        if mel.shape[0] < 10:
            raise ValueError(f"Too short: {mel.shape[0]} frames")
        
        # Convert to BigVGAN format: (1, 80, T)
        mel_t = torch.from_numpy(mel.T).unsqueeze(0).to(DEVICE)
        
        # BigVGAN inference
        with torch.no_grad():
            audio = vocoder(mel_t)
        
        # Convert to numpy
        audio = audio.squeeze().cpu().numpy()
        
        # Validate
        if np.isnan(audio).any() or np.isinf(audio).any():
            raise ValueError("Invalid audio output")
        audio = add_natural_artifacts(audio, sr=SR)
        # ===== POST-FILTER (NEW) =====
        if ENABLE_POST_FILTER:
            audio = post_filter_audio(
                audio, 
                sr=SR, 
                cutoff=CUTOFF_FREQ, 
                order=FILTER_ORDER
            )
        
        # Clip to [-1, 1]
        audio = np.clip(audio, -1.0, 1.0)
        
        # Save
        sf.write(output_path, audio, SR)
        
        success_count += 1
        
    except Exception as e:
        print(f"\n❌ Error {utt_id}: {e}")
        fail_count += 1
        continue

# ===========================
# SUMMARY
# ===========================
print("\n" + "="*70)
print("🎬 COMPLETED!")
print("="*70)
print(f"✅ Success: {success_count}/{len(mel_files)}")
print(f"❌ Failed: {fail_count}/{len(mel_files)}")
print(f"📁 Output: {OUTPUT_DIR}")

if ENABLE_POST_FILTER:
    print(f"\n🎚️  Post-filter applied:")
    print(f"   Cutoff: {CUTOFF_FREQ} Hz")
    print(f"   Order: {FILTER_ORDER}")
    print(f"   Effect: Frequencies > {CUTOFF_FREQ}Hz attenuated")

print("\n" + "="*70)
```

```python
import site, pathlib, re

fairseq_ckpt = None
for sp in site.getsitepackages():
    candidate = pathlib.Path(sp) / "fairseq" / "checkpoint_utils.py"
    if candidate.exists():
        fairseq_ckpt = candidate
        break

text = fairseq_ckpt.read_text()
print(f"weights_only hiện có: {text.count('weights_only')}")

# Patch bằng regex — không quan tâm indent
# Tìm tất cả torch.load(...) chưa có weights_only, thêm vào trước dấu ) cuối
def add_weights_only(s):
    pattern = r'(torch\.load\((?:[^()]*|\((?:[^()]*|\([^()]*\))*\))*\))'
    def replacer(m):
        call = m.group(1)
        if 'weights_only' in call:
            return call  # đã có rồi
        # Thêm weights_only=False trước ) cuối cùng
        return call[:-1] + ', weights_only=False)'
    return re.sub(pattern, replacer, s, flags=re.DOTALL)

patched = add_weights_only(text)
count_after = patched.count('weights_only=False')
print(f"weights_only=False sau patch: {count_after}")

if count_after >= 2:
    fairseq_ckpt.write_text(patched)
    print("✅ Patch thành công")
    # Verify — in ra các dòng có torch.load
    for i, line in enumerate(patched.splitlines()):
        if 'torch.load' in line or 'weights_only' in line:
            print(f"  {i+1:4d} | {line.rstrip()}")
else:
    print("❌ Regex không match đủ — thử cách khác bên dưới")
```

```python
import torch
import torchaudio
import math
from diffusers import AudioLDM2Pipeline
import librosa
import numpy as np

pipe = AudioLDM2Pipeline.from_pretrained("cvssp/audioldm2-music", torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# Load WAV gốc
orig, orig_sr = torchaudio.load("/kaggle/working/so-vits-svc/results/2044.wav_0key_AYueYue_sovits_rmvpe.wav")
orig_duration = orig.shape[-1] / orig_sr

# Detect BPM
y, sr = librosa.load("/kaggle/working/stage8_audio/2044.wav", sr=None)
tempo, _ = librosa.beat.beat_track(y=y, sr=sr)
bpm = int(round(float(tempo)))
print(f"BPM detected: {bpm}")

if bpm < 80:
    tempo_desc = f"slow ballad, {bpm} BPM, gentle, soft"
elif bpm < 120:
    tempo_desc = f"mid-tempo pop, {bpm} BPM, moderate"
else:
    tempo_desc = f"upbeat pop, {bpm} BPM, energetic"

prompt = f"""
A soft and emotional pop ballad in A minor, slow tempo around {tempo_desc}, featuring a gentle piano melody with ambient pads and light strings. The melody has a clear, singable hook with stepwise movement, natural pauses and breathing space between phrases. Chord progression follows Am–F–C–G with smooth harmonic flow. The arrangement is intimate and restrained, no dramatic climax, keeping a tender and heartfelt mood throughout. Similar to quiet pop ballads by Adele or Sam Smith. Soft percussion, warm and cinematic atmosphere.
"""
negative_prompt = "loud, noise, distortion, low quality, random"
print(f"Prompt: {prompt}")

chunk_s = 10
n_chunks = math.ceil(orig_duration / chunk_s)

chunks = []
for i in range(n_chunks):
    audio = pipe(prompt, negative_prompt=negative_prompt,
                 num_inference_steps=200,
                 audio_length_in_s=chunk_s).audios[0]
    chunks.append(torch.tensor(audio).unsqueeze(0))

gen = torch.cat(chunks, dim=-1)

gen_sr = 16000
resampler = torchaudio.transforms.Resample(gen_sr, orig_sr)
gen = resampler(gen)

if gen.shape[-1] < orig.shape[-1]:
    gen = torch.nn.functional.pad(gen, (0, orig.shape[-1] - gen.shape[-1]))
else:
    gen = gen[:, :orig.shape[-1]]

if orig.shape[0] != gen.shape[0]:
    gen = gen.expand(orig.shape[0], -1)

# Mix - nhạc nền 0.5
mixed = orig * 1.0 + gen * 0.5
mixed = mixed / mixed.abs().max()
torchaudio.save("/kaggle/working/final_mix.wav", mixed, orig_sr)
print("Done!")
```
